```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\ADITYA SINGH\Desktop\Project 1 - Mercedes-Benz Greener Manufacturing\train.csv")
df.head()
```

```python
df.drop(columns=["ID"],inplace=True)
df.head()
```

```python
def check_var(df):
    l=[]
    for i in df.columns:
        l.append(i)
    for i in range(9,len(l)):
        if df.var()[l[i]]==0:
            df.drop(columns=[l[i]], inplace=True)
            print(l[i])
check_var(df)
df.head()  
```

```python
df.isnull().values.any()
        
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()   
df["X0"]=le.fit_transform(df["X0"])
df["X1"]=le.fit_transform(df["X1"])
df["X2"]=le.fit_transform(df["X2"])
df["X3"]=le.fit_transform(df["X3"])
df["X4"]=le.fit_transform(df["X4"])
df["X5"]=le.fit_transform(df["X5"])
df["X6"]=le.fit_transform(df["X6"])
df["X8"]=le.fit_transform(df["X8"])
df.head()
```

```python
y=df.y
df.drop(columns=["y"],inplace=True)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=50)
x_pca=pca.fit_transform(x)
x_pca
```

```python
x_pca.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_pca, y, test_size = 0.25)
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
xg_reg1 = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.6, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 100)
xg_reg1.fit(x_train,y_train)
preds = xg_reg1.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test, preds))
print("RMSE: %f" % (rmse))
```

```python
xg_reg = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.6, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 100)
xg_reg.fit(x_pca,y)
```
